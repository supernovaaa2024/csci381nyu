## Introduction

**Regression** is a method used to predict a numeric value, such as house price, salary, or delivery time.
$$ y = f(x) $$
Where:

- x = input
- y = value we want to predict



#### **Linear Regression**

Linear Regression is the simplest type of regression. It assumes the relationship between input and output is a straight line.

$$ y = b_0 + b_1x $$

Where:

- $b_0$ = intercept (starting value)
- $b_1$ = slope (how much \(y\) changes when \(x\) changes)

**Advantages**
- Simple and easy to understand
- Fast to train
- Low risk of overfitting

**Disadvantages**
- Only works well when the relationship is roughly linear

---

#### **Ridge Regression (L2)**

- Adds **L2 penalty** to shrink coefficients.
- Reduces overfitting but **keeps all features**.
- Useful when predictors are correlated.
- Parameter: `alpha` (strength of regularization)
    - Small alpha → model behaves like Linear Regression
    - Large alpha → simpler model, but may underfit

#### **Lasso Regression (L1)**

- Adds **L1 penalty** to shrink some coefficients to **exactly zero**.
- Performs **feature selection** automatically.
- Good when only a few features are important.
- Parameter: `alpha` (strength of regularization)
    - Small alpha → model behaves like Linear Regression
    - Large alpha → simpler model, but may underfit

#### **ElasticNet**

- Combines **L1 + L2 penalties** (Lasso + Ridge).
- Useful when many features are correlated.
- Parameters:
  - `alpha`: overall regularization strength
  - `l1_ratio`: balance between L1 and L2
    - `0` → mostly Ridge
    - `1` → mostly Lasso

---

R² measures how well the model explains the data.
