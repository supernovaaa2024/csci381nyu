# Identifiers

```python
# . = any character, except the new line
print(re.search('a.','aaaaa')!=None)
print(re.search('a.','aa')!=None)
print(re.search('a.','ab')!=None)
print(re.search('a.','a1')!=None)
print(re.search('a.','a#')!=None)
print(re.search('a.','a')!=None)
print(re.search('a.','a\n')!=None)
```

```python
# ? = match 0 or 1 repetitions
print(re.search('ba?b','bb')!=None)
print(re.search('ba?b','bab')!=None)
print(re.search('ba?b','abab')!=None)
print(re.search('ba?b','baab')!=None)
```

```python
# * = match 0 or more repetitions
print(re.search('ba*b','bb')!=None)
print(re.search('ba*b','bab')!=None)
print(re.search('ba*b','baab')!=None)
print(re.search('ba*b','baaaaaaaab')!=None)
print(re.search('ba*b','baaaaaagb')!=None)
```

```python
# + = match 1 or more repetitions
print(re.search('ba+b','bb')!=None)
print(re.search('ba+b','bab')!=None)
print(re.search('ba+b','baab')!=None)
print(re.search('ba+b','baaaaab')!=None)
```

```python
# {} = matches a range of number of times
print(re.search('ba{1,3}b','bab')!=None)
print(re.search('ba{1,3}b','baab')!=None)
print(re.search('ba{1,3}b','baaab')!=None)
print(re.search('ba{1,3}b','baaaab')!=None)
print(re.search('ba{1,3}b','bb')!=None)
```

```python
# ^ = matches start of a string
print(re.search('^a','ab')!=None)
print(re.search('^a','abcde')!=None)
print(re.search('^a',' ab')!=None)
```

```python
# $ = mtachs end of a string
print(re.search('a$','ba')!=None)
print(re.search('a$','bcdea')!=None)
print(re.search('a$','badddd')!=None)
```

```python
# Example-1
print(re.search('^a.a$','aba')!=None)
print(re.search('^a.a$','a a')!=None)
print(re.search('^a.a$','abbba')!=None)
```

```python
# Example-2
print(re.search('^a.?a$','aba')!=None)
print(re.search('^a.?a$','aa')!=None)
```

```python
# Example-3
print(re.search('^a.*a$','aba')!=None)
print(re.search('^a.*a$','a        a')!=None)
print(re.search('^a.*a$','abcdefgha')!=None)
```

```python
# Example-4
print(re.search('^a.{1,3}a$','aba')!=None)
print(re.search('^a.{1,3}a$','abcda')!=None)
print(re.search('^a.{1,3}a$','abcdefa')!=None)
```

```python
# [] are used to specify a set of characters to match
print(re.search('[123abc]','hi')!=None)
print(re.search('[123abc]','welcome')!=None)
print(re.search('[123abc]','1abc')!=None)
print(re.search('[123abc]','hi2')!=None)
print(re.search('[1-3a-c]','welcome')!=None)
```

```python
# () are used to determine a group of characters (together)
print(re.search('(abc)','abc')!=None)
print(re.search('(abc){2,3}','abc')!=None)
print(re.search('(abc){2,3}','abcabc')!=None)
print(re.search('(abc){2,3}','abcabcabc')!=None)
print(re.search('(abc){2}','abcabcabc')!=None)
print(re.search('^(abc){2,3}$','abc')!=None)
print(re.search('^(abc){2,3}$','abcabc')!=None)
print(re.search('^(abc){2,3}$','abcabcabc')!=None)
```

```python
# | vertical baris a logical or
print(re.search('abc|123','a')!=None)
print(re.search('abc|123','1')!=None)
print(re.search('abc|123','123')!=None)
print(re.search('abc|123','abc')!=None)
```

```python
# \ backslash is used with special characters
print(re.search('a?','123')!=None)   # seacrh for a 0 or 1 time
print(re.search('a\?','123')!=None)  # seacrh for a?
```

```python
# Special Characters
# \d = matches any decimal digit [0-9]
print(re.search('\d','welcome123')!=None)
print(re.search('\d','123')!=None)
print(re.search('\d','welcome')!=None)
```

```python
# \D = matches any non-digit  character
print(re.search('\D','welcome123')!=None)
print(re.search('\D','welcome')!=None)
print(re.search('\D','123')!=None)
```

```python
# \w = matches any alphanumeric character
print(re.search('\w','welcome123')!=None)
print(re.search('\w','welcome')!=None)
print(re.search('\w','123')!=None)
print(re.search('\w','welcome123.,?')!=None)
print(re.search('\w','.?#')!=None)
```

```python
# \W = matches any non-alphanumeric character
print(re.search('\W','welcome123')!=None)
print(re.search('\W','welcome')!=None)
print(re.search('\W','123')!=None)
print(re.search('\W','welcome123.,?')!=None)
print(re.search('\W','.?#')!=None)
```

```python
# \s = matches any white space character
print(re.search('\s','welcome every one')!=None)
print(re.search('\s','welcome')!=None)
print(re.search('\s','welcome\teveryone')!=None)
```

```python
# \S = matches any non-white space character
print(re.search('\S','welcome every one')!=None)
print(re.search('\S','welcome')!=None)
print(re.search('\S','welcome\teveryone')!=None)
print(re.search('\S',' \t\n')!=None)
```
