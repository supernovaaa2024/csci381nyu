## 5. Listing & Viewing Files

```python
ls        # List files and directories in the current directory
ls -l     # List files and directories in long format (permissions, owner, size, date)
ls -a     # List all files, including hidden files (those starting with .)
ls -la    # Long format listing including hidden files
```

### cat / less / head / tail / stat

```python
echo -e "hi\nwelcome"
```

```python
echo -e "a\nb\nc\nd" > sample.txt
```

```python
cat sample.txt        # Display the entire contents of the file in the terminal
nl sample.txt         # Display the file contents with line numbers added
head -n 2 sample.txt  # Show only the first 2 lines of the file
tail -n 2 sample.txt  # Show only the last 2 lines of the file
tail -n +2 sample.txt # Show the file starting from line 2 to the end
stat sample.txt       # Display detailed file metadata (size, permissions, timestamps, inode)
```

## 6. Creating Files & Directories

```python
mkdir testdir                 # Create a new directory named testdir
mkdir -p testdir/sub1/sub2    # Create nested directories; no error if they already exist
touch file1.txt               # Create an empty file (or update timestamp if it exists)
```

## 7. Navigation

```python
pwd     # Print the full path of the current working directory
cd ..   # Move to the parent directory
cd ~    # Change to the current user's home directory
cd /
cd /folder1/folder2
cd ./folder1/folder2
```

## 8. Copying, Moving, Deleting

```python
cp file1.txt file2.txt        # Copy file1.txt to a new file named file2.txt
cp -r testdir testdir_copy    # Recursively copy the directory testdir and its contents
mv file2.txt renamed.txt     # Rename (or move) file2.txt to renamed.txt
rm renamed.txt               # Delete the file renamed.txt
rm -r testdir_copy           # Recursively delete the directory testdir_copy and its contents
```

## 9. Searching & Finding

```python
find . -name "*.txt"   # Find all .txt files starting from the current directory
find . -type f         # Find all regular files starting from the current directory
find . -type d         # Find all directories starting from the current directory
```

```python
grep "ERROR" sample.txt          # find all lines containing "ERROR"
grep -i "error" sample.txt        # case-insensitive search
grep -c "ERROR" sample.txt        # count matching lines
grep -n "ERROR" sample.txt        # show line numbers of matches
grep -v "INFO" sample.txt         # show lines NOT containing "INFO"
grep -r "TODO" ./data/            # recursive search in directory
grep -w "ERROR" sample.txt        # match whole word only
```

# 10. Redirection & Pipes

```python
ls > files.txt            # Redirect ls output to files.txt (overwrite if it exists)
ls >> files.txt           # Append ls output to the end of files.txt
wc < files.txt            # Count lines, words, and bytes from files.txt via input redirection
cat files.txt | wc -l     # Count the number of lines in files.txt using a pipe
```

## 11. wc (Word Count)

```python
wc sample.txt        # Show number of lines, words, and bytes in sample.txt
wc -l sample.txt     # Show number of lines in sample.txt
wc -w sample.txt     # Show number of words in sample.txt
wc -c sample.txt     # Show number of bytes in sample.txt
```

## 12. sort

```python
echo -e "3\n1\n2" > nums.txt   # Create nums.txt with three lines: 3, 1, 2
```

```python
sort nums.txt                  # Sort the lines alphabetically (lexicographically)
sort -r nums.txt               # Sort the lines in reverse order
sort -n nums.txt               # Sort the lines numerically
sort -b nums.txt               # Sort lines ignoring leading spaces
```

## 13. cut

```python
echo -e "name,age,salary\nsarah,20,2000\njames,25,5000" > cut.txt         # Create cut.txt with the text "a:b:c:d"
```

```python
cat cut.txt
```

```python
cut -d, -f3 cut.txt
```

```python
cut -c1-3 cut.txt                # Extract characters 1 to 3 from each line
cut -d: -f1 cut.txt              # Extract the first field using ':' as the delimiter
cut -d: -f1,3 cut.txt            # Extract the first and third fields separated by ':'
cut -d: -f-3 cut.txt             # Extract fields 1 through 3 (inclusive) separated by ':'
cut -d: -f1-3 cut.txt            # Another way to extract fields 1 through 3 separated by ':'
cut -b1-4 cut.txt                # Extract bytes 1 to 4 from each line
```

## 14. uniq

```python
echo -e "a\na\nb\nb\nc" > uniq.txt   # Create uniq.txt with repeated letters
```

```python
cat uniq.txt
```

```python
sort uniq.txt | uniq                   # Remove duplicate lines (show unique lines)
sort uniq.txt | uniq -c                # Count occurrences of each line
sort uniq.txt | uniq -d                # Show only lines that are repeated
sort uniq.txt | uniq -u                # Show only lines that appear once
```
