# re

```python
# Import the regular expressions module for pattern matching
import re
```

```python
# Define a sample text string
txt = "Hi, how are you today?"
# Search for the substring 'Hi' in txt using regular expressions
re.search('Hi', txt)
```

```python
# Find the substring 'Hi' in txt using the built-in string method
txt.find('Hi')
```

```python
# Use re.search to find 'Hi' in txt and store the match object in s
s = re.search('Hi', txt)
# Check if the match object is not None (i.e., a match was found)
s != None
```

# Identifiers

```python
# . = any character, except the new line
# Checks if there is an 'a' followed by any character
print(re.search('a.','aaaaa')!=None)  # True: matches 'aa'
print(re.search('a.','aa')!=None)     # False: no char after 'a'
print(re.search('a.','ab')!=None)     # True: matches 'ab'
print(re.search('a.','a1')!=None)     # True: matches 'a1'
print(re.search('a.','a#')!=None)     # True: matches 'a#'
print(re.search('a.','a')!=None)      # False: no char after 'a'
print(re.search('a.','a\n')!=None)   # False: . does not match newline
print(re.search('a.','a ')!=None)     # True: matches 'a '
```

```python
# ? = match 0 or 1 repetitions
# 'a' is optional between 'b' and 'b'
print(re.search('ba?b','bb')!=None)    # True: matches 'bb'
print(re.search('ba?b','bab')!=None)   # True: matches 'bab'
print(re.search('ba?b','abab')!=None)  # True: matches 'bab' in 'abab'
print(re.search('ba?b','baab')!=None)  # False: two 'a's, not allowed
print(re.search('ba?b','batb')!=None)  # False: 't' is not optional
```

```python
# * = match 0 or more repetitions
# 'a' can appear zero or more times between 'b' and 'b'
print(re.search('ba*b','bb')!=None)           # True: zero 'a'
print(re.search('ba*b','bab')!=None)          # True: one 'a'
print(re.search('ba*b','baab')!=None)         # True: two 'a's
print(re.search('ba*b','baaaaaaaab')!=None)   # True: many 'a's
print(re.search('ba*b','baaaaaagb')!=None)    # False: 'g' breaks the pattern
```

```python
# + = match 1 or more repetitions
# 'a' must appear at least once between 'b' and 'b'
print(re.search('ba+b','bb')!=None)        # False: no 'a'
print(re.search('ba+b','bab')!=None)       # True: one 'a'
print(re.search('ba+b','baab')!=None)      # True: two 'a's
print(re.search('ba+b','baaaaab')!=None)   # True: many 'a's
```

```python
# {} = matches a range of number of times
# 'a' must appear between 1 and 3 times between 'b' and 'b'
print(re.search('ba{1,3}b','bab')!=None)      # True: 1 'a'
print(re.search('ba{1,3}b','baab')!=None)     # True: 2 'a's
print(re.search('ba{1,3}b','baaab')!=None)    # True: 3 'a's
print(re.search('ba{1,3}b','baaaab')!=None)   # False: 4 'a's
print(re.search('ba{1,3}b','bb')!=None)       # False: 0 'a'
```

```python
# ^ = matches start of a string
# Checks if string starts with 'a'
print(re.search('^a','ab')!=None)      # True: starts with 'a'
print(re.search('^a','abcde')!=None)   # True: starts with 'a'
print(re.search('^a',' ab')!=None)     # False: starts with space
```

```python
# $ = matches end of a string
# Checks if string ends with 'a'
print(re.search('a$','ba ')!=None)     # False: ends with space
print(re.search('a$','bcdea')!=None)   # True: ends with 'a'
print(re.search('a$','badddd')!=None)  # False: does not end with 'a'
```

```python
# Example-1
# Matches any 3-character string that starts and ends with 'a'
print(re.search('^a.a$','aba')!=None)      # True: 'aba'
print(re.search('^a.a$','a a')!=None)      # True: 'a a'
print(re.search('^a.a$','abbba')!=None)    # False: too long
```

```python
# Example-2
# Matches string that starts and ends with 'a', with 0 or 1 character in between
print(re.search('^a.?a$','abba')!=None)    # False: too many chars in between
print(re.search('^a.?a$','aa')!=None)      # True: no char in between
```

```python
# Example-3
# Matches string that starts and ends with 'a', with any number of characters in between
print(re.search('^a.*a$','aba')!=None)         # True: 'aba'
print(re.search('^a.*a$','a        a')!=None)  # True: spaces in between
print(re.search('^a.*a$','abcdefgha')!=None)   # True: 'abcdefgha'
```

```python
# Example-4
# Matches string that starts and ends with 'a', with 1 to 3 characters in between
print(re.search('^a.{1,3}a$','aba')!=None)         # True: 1 char in between
print(re.search('^a.{1,3}a$','abcda')!=None)       # True: 3 chars in between
print(re.search('^a.{1,3}a$','abcdefa')!=None)     # False: 5 chars in between
```

```python
# [] are used to specify a set of characters to match
print(re.search('[123abc]','hi')!=None)         # False: no match
print(re.search('[123abc]','welcome')!=None)    # True: 'c' matches
print(re.search('[123abc]','1abc')!=None)       # True: '1' matches
print(re.search('[123abc]','hi2')!=None)        # True: '2' matches
print(re.search('[1-3a-c]','welcome')!=None)    # True: 'c' matches (range)
```

```python
# () are used to group characters together
print(re.search('(abc)','abc')!=None)               # True: matches 'abc'
print(re.search('(abc){2,3}','abc')!=None)          # False: only one 'abc'
print(re.search('(abc){2,3}','abcabc')!=None)       # True: two 'abc'
print(re.search('(abc){2,3}','abcabcabc')!=None)    # True: three 'abc'
print(re.search('(abc){2}','abcabcabc')!=None)      # True: at least two 'abc'
print(re.search('^(abc){2,3}$','abc')!=None)        # False: only one 'abc'
print(re.search('^(abc){2,3}$','abcabc')!=None)     # True: two 'abc'
print(re.search('^(abc){2,3}$','abcabcabc')!=None)  # True: three 'abc'
```

```python
# | vertical bar is a logical OR
print(re.search('abc|123','a')!=None)       # False: neither 'abc' nor '123'
print(re.search('abc|123','1')!=None)       # False: neither 'abc' nor '123'
print(re.search('abc|123','123')!=None)     # True: matches '123'
print(re.search('abc|123','abc')!=None)     # True: matches 'abc'
```

```python
# \ backslash is used with special characters
print(re.search('a?','123')!=None)    # True: 'a' zero or one time (always matches)
print(re.search('a\\?','123')!=None)  # False: looking for literal 'a?'
```

```python
# \\d matches any digit (0-9)
print(re.search(r'a\\d','a1')!=None)    # True: 'a' followed by a digit
print(re.search(r'a\\d','a')!=None)     # False: no digit after 'a'
print(re.search(r'a\\d','1a')!=None)    # False: 'a' is not followed by a digit
```

```python
# \\D matches any non-digit character
print(re.search(r'a\\D','a1')!=None)    # False: 'a' is followed by a digit
print(re.search(r'a\\D','ab')!=None)    # True: 'a' is followed by a non-digit
```

```python
# \\s matches any whitespace character (space, tab, newline)
print(re.search(r'a\\s','a ')!=None)    # True: 'a' followed by a space
print(re.search(r'a\\s','a\t')!=None)  # True: 'a' followed by a tab
print(re.search(r'a\\s','ab')!=None)    # False: 'a' is not followed by whitespace
```

```python
# \\S matches any non-whitespace character
print(re.search(r'a\\S','a ')!=None)    # False: 'a' is followed by a space
print(re.search(r'a\\S','ab')!=None)    # True: 'a' is followed by a non-whitespace character
```

```python
# \\w matches any alphanumeric character (a-z, A-Z, 0-9, and underscore)
print(re.search(r'a\\w','a_')!=None)    # True: 'a' followed by an alphanumeric or underscore
print(re.search(r'a\\w','a-')!=None)    # False: '-' is not alphanumeric or underscore
```

```python
# \\W matches any non-alphanumeric character
print(re.search(r'a\\W','a_')!=None)    # False: '_' is alphanumeric or underscore
print(re.search(r'a\\W','a-')!=None)    # True: '-' is not alphanumeric or underscore
```

```python
# Example: match a phone number pattern (3 digits, dash, 3 digits, dash, 4 digits)
print(re.search('\d{3}-\d{3}-\d{4}','123-456-7890')!=None)    # True: matches phone number format
print(re.search('\d{3}-\d{3}-\d{4}','123-45-7890')!=None)     # False: does not match format
```

### Some functions

```python
# Example: match an email address pattern (simple version)
print(re.search('[\w.-]+@[\w.-]+\.\w+','test@example.com')!=None)    # True: matches email format
print(re.search('[\w.-]+@[\w.-]+\.\w+','test.example.com')!=None)     # False: does not match email format
```

```python
# Example: match a date in the format YYYY-MM-DD
print(re.search('\d{4}-\d{2}-\d{2}','2023-05-01')!=None)    # True: matches date format
print(re.search('\d{4}-\d{2}-\d{2}','23-05-01')!=None)       # False: does not match format
```

```python
# Example: match a US ZIP code (5 digits, optionally followed by a dash and 4 digits)
print(re.search('^\d{5}(-\d{4})?$','12345')!=None)        # True: matches 5-digit ZIP
print(re.search('^\d{5}(-\d{4})?$','12345-6789')!=None)  # True: matches ZIP+4
print(re.search('^\d{5}(-\d{4})?$','1234')!=None)         # False: not enough digits
```

```python
# Example: match a hexadecimal color code (e.g., #A1B2C3)
print(re.search('^#[0-9A-Fa-f]{6}$','#A1B2C3')!=None)    # True: matches hex color
print(re.search('^#[0-9A-Fa-f]{6}$','A1B2C3')!=None)     # False: missing '#'
```
