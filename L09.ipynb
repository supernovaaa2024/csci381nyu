We can create DataFrames using dictionaries or lists of lists.

```python
import pandas as pd

# Create df1 with 'id' values [10, 20]
df1 = pd.DataFrame({
    'id': [10, 20],
    'data_x': ['A', 'B']
})

# Create df2 with 'id' values [20, 30]
df2 = pd.DataFrame({
    'id': [20, 30],
    'data_y': ['C', 'D']
})

# Perform an inner merge on the 'id' column
result = pd.merge(df1, df2, on='id', how='inner')
concat = pd.concat([df1, df2], ignore_index=True, axis=0, join='inner')

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
print("\nMerge Result:")
print(result)
print("\nNumber of rows in result:", len(result))
print("\nConcatenated Result:")
print(concat)
```

### 7. Pivot Table Explanation

A pivot table in Pandas is used to **summarize and aggregate data** in a table format.

- `index` → rows
- `columns` → columns
- `values` → values to aggregate
- `aggfunc` → aggregation function (default is `mean`)

It is especially useful for **grouping data** and **computing statistics** across multiple categories.


```python
# Basic pivot table: average age per education and department
df.pivot_table(index='education', columns='department', values='age')  

# Pivot table with sum aggregation
df.pivot_table(index='education', columns='department', values='age', aggfunc='sum')  

# Pivot table with multiple aggregation functions
df.pivot_table(index='education', columns='department', values='age', aggfunc=['mean','min','max'])  

# Pivot table including all columns
df.pivot_table(index='education', values=['age','id'], aggfunc='mean')  

# Pivot table with fill value for missing data
df.pivot_table(index='education', columns='department', values='age', fill_value=0)  
```

To achieve the exact same result as `df.loc['MBA', ['age', 'name']]` without using `.loc`, you have to split the operation into two chained steps: one for the rows, and one for the columns.

Here is the most common equivalent using just standard brackets `[]`:

**`df[df.index == 'MBA'][['age', 'name']]`**

Here is exactly what that does step-by-step:

1. **`df[df.index == 'MBA']`**: This filters the rows. Because "MBA" is in the index of your DataFrame (the bolded column on the far left), you use `df.index == 'MBA'` to grab only those rows.
2. **`[['age', 'name']]`**: This takes the result of the first step and filters it down to only the two columns you want.

You can also flip the order and it will do the exact same thing:
**`df[['age', 'name']][df.index == 'MBA']`**

**Why `.loc` is still better here:**
While the chained bracket method above works perfectly for *viewing* the data, pandas strongly recommends using `.loc` for this specific operation. If you ever try to assign new values using the chained method (e.g., `df[df.index == 'MBA'][['age', 'name']] = 30`), pandas will throw a `SettingWithCopyWarning` because it gets confused about whether you are modifying the original DataFrame or a temporary copy. `.loc` handles this flawlessly in a single step!
