### Question 1

There are 12 files in the folder `sales-data`: (1.txt, 2.txt, 3.txt, .. 12.txt). Each one of these files contains sales amount of three different items (item1, item2 items) in 30 days (each line represents one day). 

**a.**  Write python code to create a new file `totals.txt` that contains the total sale of each item in each month. The content of the file should look like this:

```
item0,item1,item2
164714,171778,153696
162723,181867,166487
172914,149516,155791
179961,153297,173636
159424,174871,149985
181047,157718,155231
153844,155137,181145
175812,175479,169075
177620,167257,167605
180829,178063,170401
172265,176444,152275
151173,159185,139130
```

````python
## Your code here
# python code to create a new file `totals.txt` that contains the total sale of each item in each month
with open("totals.txt", "w") as output_file: # Creates the new file
    output_file.write("item0,item1,item2\n")
    for month in range(1,13):
        total_item0 = 0 
        total_item1 = 0
        total_item2 = 0
        with open(f"sales-data/{month}.txt", "r") as input_file:
            lines = input_file.readlines()
            for line in lines[1:]:
                line = line.strip()
                if line: # runs if line is not empty
                    parts = line.split(",")
                    total_item0 += int(parts[0])
                    total_item1 += int(parts[1])
                    total_item2 += int(parts[2])
        output_file.write(f"{total_item0},{total_item1},{total_item2}\n")
````

**b.** use the created file `totals.txt` to print a list of items names sorted by their total sales (form largest to smallest). The output should look like this:

```
['item0', 'item1', 'item2']
```

````python
try:
    with open("totals.txt", "r") as f:
        lines = [line.strip() for line in f if line.strip()]
except FileNotFoundError:
    print("totals.txt not found. Run Q1(a) first.")
else:
    items = lines[0].split(",")

    totals = [0] * len(items)
    for row in lines[1:]:
        values = row.split(",")
        for i in range(len(items)):
            totals[i] += int(values[i])

    sorted_items = [name for name, total in sorted(zip(items, totals), key=lambda x: x[1], reverse=True)]
    print(sorted_items)
````

### Question 2.
 
**a.** Using Data Structure(s), create a Python representation of a book that includes the following information:
- The following attributes:
    - `title`
    - `authors_lastnames` (list of authors_lastnames)
    - `publisher_name`
    - `publishing_year`

- Also, provide a function named `get_citation(book)` that takes a book as input returns the book’s citation as a formatted string in the form:

    ```authors_lastnames (publishing_year). title, publisher.```
    
For example:
    `Meysman, Cielen (2016). Introducing Data Science, manning`

**Notes:** 
- If there is more than one author be sure to add `,` between their lastnames.


````python
## Your code here
def create_book(title, authors_lastnames, publisher_name, publishing_year):
    return {
        "title": title,
        "authors_lastnames": authors_lastnames,
        "publisher_name": publisher_name,
        "publishing_year": publishing_year
    }
def get_citation(book):
    authorsjoined = ", ".join(book["authors_lastnames"])   
    citation = f"{authorsjoined} ({book['publishing_year']}). {book['title']}, {book['publisher_name']}." # The Professor says we could add or not add fullstop here both is fine.
    return citation
````

**b.** Use the created structure to create a list of books call it `list_of_books`  using the data in file `booksdata.txt`

````python
## Your code here
# Create The Books
list_of_books = []
with open("booksdata.txt","r") as f:
    lines = f.readlines()
    for line in lines:
        if line:
            parts = line.strip().split(",",3) # Split into 4 parts by limitting the number of splits to 3 since the last part could have commas.

            authors_lastnames = parts[0].split("|")
            publishing_year = parts[1]
            title = parts[2]
            publisher_name = parts[3]

            book = create_book(title, authors_lastnames, publisher_name, publishing_year)
            list_of_books.append(book)
````

**c.** Loop through the books in the `list_of_books` then find the latest 3 books (recently published) and show their citation.

The output should be like this:

Robinson, Nolis (2020). Build A Career In Data Science, Manning
Carter (2020). Data Science For Mathematicians, CRC Press

````python
## Your code here
sorted_books = sorted(list_of_books, key=lambda book: int(book["publishing_year"]), reverse=True)
latest3 = sorted_books[:3] # gets first 3
for book in latest3:
    print(get_citation(book).rstrip("."))
# The example output has no fullstop at the end so I added rstrip(".") to remove the fullstop at the end of the citation.
# Professor says with or without fullstop is fine, so I just removed it to match the example output.
````
