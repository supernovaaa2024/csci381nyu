```python
# read the content of the file ../data/chat.txt
import pandas as pd
pd.read_csv('chat.txt', header=None, names=['text'])
```

```python
# find the total number of participants in this chat
import re
with open('chat.txt', 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]
header_lines = lines[::2]
participants = [re.sub(r'\s\d{2}:\d{2}:\d{2}$', '', line) for line in header_lines]
# participants = [re.findall(r'^(.+)\s\d{2}:\d{2}:\d{2}$', line)[0] for line in header_lines]
total_participants = len(set(participants))
total_participants
print(f"Total number of participants: {total_participants}")
```

```python
# find the one who partipated most
from collections import Counter
message_count = Counter(participants)
top_participant, top_messages = message_count.most_common(1)[0]
top_participant, top_messages
```

```python
# find the duration of the chat
import re
from datetime import datetime
with open('chat.txt', 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]
header_lines = lines[::2]
times = [re.search(r'(\d{2}:\d{2}:\d{2})$', line).group(1) for line in header_lines]
start_time = datetime.strptime(times[0], '%H:%M:%S')
end_time = datetime.strptime(times[-1], '%H:%M:%S')
chat_duration = end_time - start_time
chat_duration
```

```python
# find the topic these people were talking about (i.e find the most frequent_words)
import re
from collections import Counter
with open('chat.txt', 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]
messages = lines[1::2]
text = ' '.join(messages).lower()
words = re.findall(r"[a-z']+", text)
stop_words = {
    'the','a','an','and','or','to','of','in','on','at','for','with','is','are','was','were',
    'be','been','being','this','that','it','as','by','from','under','over','throughout',
    'will','can','how','when','why','what','who','my','your','our','their','any','every'
}
filtered_words = [w for w in words if len(w) > 2 and w not in stop_words]
top_words = Counter(filtered_words).most_common(10)
top_words
```

```python
# summary
summary = {
    'total_participants': total_participants,
    'top_participant': top_participant,
    'top_messages': top_messages,
    'chat_duration': str(chat_duration),
    'top_words': top_words[:5]
}
summary
```
