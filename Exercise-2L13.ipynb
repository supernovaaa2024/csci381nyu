```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
df=pd.read_csv('movies.csv')
df_nonull = df.dropna()
df_nonull.head()
```

```python
# find how many different languages do we have in this dataset
print(df['language'].nunique())
# Bar Chart
language_counts = df['language'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=language_counts.index, y=language_counts.values, palette='viridis')
plt.xlabel('Language')
plt.ylabel('Count')
plt.title('Count of Movies by Language')
plt.xticks(rotation=45)
plt.show()
```

```python
# show how imdb_score is distributed in this dataset (show the histogram)
plt.hist(df['imdb_score'], bins=1000, edgecolor='black')
plt.show()
```

```python
# show the histogram of the movies budgets
# import numpy as np
# plt.hist(np.log1p(df_nonull['budget']), bins=1000, edgecolor='black')
# plt.xlabel('Log(Budget)')
# plt.ylabel('Frequency')
# plt.title('Histogram of Log Movie Budgets')
# plt.show()

# change range of x-axis to 0-100 million
filtered_df = df_nonull[df_nonull['budget'] < 0.2e9]
plt.hist(filtered_df['budget'], bins=1000, edgecolor='black')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.title('Histogram of Movie Budgets')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=filtered_df['budget'], showfliers=True)  # Box plot
sns.stripplot(y=filtered_df['budget'], color='red', alpha=0.5)  # Overlay individual points
plt.title("Box Plot of Movie Budgets")
plt.show()
```

```python
# Find out if there is a relation between gross income and the budget
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['gross'], filtered_df['budget'])
plt.xlabel('Gross Income')
plt.ylabel('Budget')
plt.title('Gross Income vs Budget')
plt.xlim(0, 1000000000)  # Set x-axis limit to 1 billion
plt.ylim(0, 100000000)   # Set y-axis limit to 100 million
plt.show()
```

```python
# Find out if there is a relation between gross income and the budget
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['budget'], filtered_df['gross'], alpha=0.5)
plt.xlabel('Budget')
plt.ylabel('Gross Income')
plt.title('Budget vs Gross Income')
plt.xlim(0, 100000000)  # Set x-axis limit to 100 million
plt.ylim(0, 1000000000) # Set y-axis limit to 1 billion
plt.show()
# correlation coefficient
correlation = filtered_df['budget'].corr(filtered_df['gross'])
print(f"Correlation coefficient between budget and gross income: {correlation:.2f}")
```
