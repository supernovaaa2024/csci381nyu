## **Lambda vs Function in Python**

In Python, both **lambda expressions** and **functions** are used to define reusable pieces of code. However, they differ in syntax, capabilities, and typical use cases.


- A regular function is defined using the `def` keyword. It can contain multiple statements and supports complex logic.
- A `lambda` function is an anonymous, single-expression function defined using the lambda keyword.


```python
def square(x):              # define a function that returns the square of x
    return x * x            # multiply x by itself

square_lambda = lambda x: x * x   # define a lambda function to square x

square(5)                   # call the square function with argument 5
square_lambda(5)            # call the lambda function with argument 5
```

```python
def square(x):
    return x*x

square_lambda = lambda x: x*x

print(square(5))
print(square_lambda(5))
```

## **List Comprehensions**

A **list comprehension** is a concise and readable way to create lists in Python.  
It allows you to generate a new list by applying an expression to each element of an iterable (such as a list, range, or string), optionally filtering elements with a condition.

```python
L=[]
for x in range(10):
    L.append(x)
```

```python
L=[]
for x in range(10):
    if(x%2==0):
        L.append(x*2)
```

```python
L=[]
for x in range(10):
    if(x%2==0):
        L.append(x)
    else:
        L.append(x*2)
```

```python
[x for x in range(10)]                          # list comprehension of numbers 0–9
[x*2 for x in range(10) if x % 2 == 0]            # list of even numbers from 0–9
[x if x % 2 == 0 else x * 2 for x in range(10)] # even numbers unchanged, odd numbers doubled
```

## **Iterable vs Iterator**

- An **iterator** is an object in Python that allows you to traverse through a collection **one element at a time**. Unlike lists or tuples, iterators do not store all elements in memory at once. Instead, they generate values **on demand**, which makes them memory-efficient.

- An **iterable** is any object that can return an iterator. Examples include:
    - `list`
    - `tuple`
    - `string`
    - `set`
    - `dictionary`

```python
names = ['hasan', 'alma', 'maya', 'sarah']  # define a list of names
i = iter(names)                             # create an iterator from the list
next(i)                                    # get the first element from the iterator
list(i)                                    # convert remaining elements to a list
```

```python
next(i)
```

```python
list(i)
```

### **Why this matters in data science**

Iterators matter because they let data scientists trade memory for control and scalability.

| Situation in Data Science            | List (Iterable)                     | Iterator (Why it matters)                 |
|-------------------------------------|-------------------------------------|-------------------------------------------|
| Very large dataset                  | Loads everything into memory        | Processes data one item at a time         |
| Streaming / online data             | Not suitable                        | Natural fit (one-pass processing)         |
| Early stopping (threshold met)      | Still processes all data            | Stops immediately, saves computation      |
| Memory-constrained environment      | May crash or slow down              | Uses constant memory                      |
| Feature pipelines / generators      | Requires storing intermediate data | Enables lazy, efficient pipelines         |
| Real-time or sequential processing  | Inefficient                         | Designed for this use case                |



#### **Example-1**

```python
# Create a huge list
numbers = list(range(1_000_000_000))  # Takes a lot of memory!
```

```python
# Process first 5 numbers
for i in numbers[:5]:
    print(i)
```

```python
# Create an iterator
numbers_iter = iter(range(1_000_000_000))  # Uses almost no memory

# Process first 5 numbers
for _ in range(5):
    print(next(numbers_iter))
```

## **Generator**

- A generator is created using a function that contains the `yield` keyword.  
- When a generator function is called, it returns a **generator object** without executing the function body immediately.
- Each time the generator is iterated over (for example, using `next()` or a loop), execution resumes from where it last stopped.
- Generators are useful when working with **large datasets** or **infinite sequences** because they:
    - Save memory (values are generated on demand)
    - Improve performance for large iterations
    - Allow lazy evaluation (Compute only when needed).


| Regular Function | Generator Function |
|------------------|-------------------|
| Uses `return`    | Uses `yield`      |
| Executes fully   | Pauses and resumes |
| Returns one value | Returns a sequence of values |
| Stores all data in memory | Generates values on demand |

```python
# Regular Function (does unnecessary work)

def powers_v1(lst):           # define a function to compute squares of list elements
    p = []                    # initialize an empty list to store results
    for v in lst:             # iterate over each value in the input list
        p.append(v**2)        # append the square of the value to the result list
    return p                  # return the list of squared values

# powers_v1(L)                  # call the function with list L
max(memory_usage((powers_v1, (L,))))
```

```python
# data-science friendly

def powers_v2(lst):        # define a generator function to compute squares lazily
    for v in lst:          # iterate over each value in the input list
        yield v**2         # yield the square of the current value

g = powers_v2(L)           # create a generator object from the list L
```
