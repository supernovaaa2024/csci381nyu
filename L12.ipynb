```python
import requests
from bs4 import BeautifulSoup
content= requests.get('https://www.usatoday.com/').text
soup=BeautifulSoup(content)
news=[t.text for t in soup.find_all('a',{'class':'gnt_m_th_a'})]
f=open('news.txt', 'a')
f.write('\n'.join(news))
f.close()
```

```python
news=[t.text for t in soup.find_all('a',{'class':'gnt_m_th_a'})]
links=[t['href'] for t in soup.find_all('a',{'class':'gnt_m_th_a'})]

print('\n'.join(news))
print('----------------------------------')
print('\n'.join(links))
```

```python
keywords=' '.join(news).lower().replace('.','').replace(',','').split()
d={}
for w in keywords:
    if w not in d:
        d[w]=1
    else:
        d[w]+=1
```
