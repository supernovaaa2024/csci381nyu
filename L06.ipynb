## <u>**Text Files**</u>

Python’s built-in `open()` function is used for reading plain text files.
It is best suited for **small to medium-sized unstructured data**.

Key ideas:
- Files are read line-by-line or all at once
- Use `with` to safely manage resources


#### **Writing**

```python
# Create a sample text file
with open("sample.txt", "w") as f:
    f.write("Alice,85\nBob,90\nCharlie,78")
    
f=open("sample.txt", "w")
f.write("Alice,85\nBob,90\nCharlie,78")
f.close()
```

```python
# Print the contents of sample.txt
with open("sample.txt", "r") as f:
    print(f.read())
```


## <u>**CSV Files**</u>

CSV is one of the most common formats for structured data.
Python provides:
- `csv` module (low-level)
- `pandas.read_csv()` (high-level, recommended)


#### **Creating**

```python
import pandas as pd

# Create a sample CSV file
df = pd.DataFrame({
    "student": ["Alice", "Bob", "Charlie"],
    "score": [85, 90, 78]
})
df.to_csv("students.csv", index=False)
```

#### **Reading**

```python
# Read CSV using pandas
students = pd.read_csv("students.csv")
students
```

```python
# For large files, loading everything at once may be inefficient.
# Chunking allows processing data piece by piece.
# Read CSV in chunks
chunks = pd.read_csv("students.csv", chunksize=1)
for chunk in chunks:
    print(chunk)
```


## <u>**JSON Files**</u>

JSON is commonly used for:
- Configuration files
- API responses

Python reads JSON into dictionaries and lists.

if test we could use f.read(), f.readlines()
np.loadtxt reads file into numpy array
JSON is best opened with  json.load() - loads into Python dict
JSON is semi-structured

#### **Writing**

```python
import json

# Create a JSON file
data = {
    "course": "Data Science",
    "students": 30,
    "online": True
}

with open("course.json", "w") as f:
    json.dump(data, f)
```

#### **Reading**

```python
# Read JSON file
with open("course.json", "r") as f:
    course_data = json.load(f)

course_data
```


## <u>**Databases**</u>

SQLite is a lightweight database stored in a file.
Useful for small applications and teaching SQL basics.


#### **Writing**

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

cursor.execute("CREATE TABLE IF NOT EXISTS students (name TEXT, score INTEGER)")
cursor.execute("INSERT INTO students VALUES ('Alice', 85), ('Bob', 90), ('Charlie', 78)")
conn.commit()
import os
print(os.getcwd())
```

#### **Reading**

```python
# Read data from SQL into pandas
import pandas as pd
query = "SELECT * FROM students"
df_sql = pd.read_sql_query(query, conn)
df_sql
```
