## 3. Inspection and Preparation
Understanding the metadata and structure of your data.

```python
print("Index:", df1.index)
print("Columns:", df1.columns)
print("Values:\n", df1.values)

df1.info()      # Structure and data types
df1.describe()  # Statistical summary

# Index Manipulation
df_indexed = df1.set_index('name')
df_reset = df_indexed.reset_index()

# Dropping Data
df_no_salary = df1.drop('salary', axis=1) # Drop Column
df_no_row = df1.drop(0, axis=0)           # Drop first row
```

## 4. Selecting Data
Methods for accessing specific rows and columns.

```python
# Selection via column and row index
print("Single value [col][row]:", df1['name'][1])

# Label-based selection (.loc)
print("Loc selection:\n", df_indexed.loc['maya', 'age'])

# Position-based selection (.iloc)
print("Iloc selection (1st row, 2nd col):", df1.iloc[0, 1])
```

## 5. Handling Missing Values
Common strategies for cleaning incomplete data.

```python
df_na = pd.DataFrame({'A': [1, np.nan, 3], 'B': [np.nan, 5, 6]})

print("Is NA Check:\n", df_na.isna())
print("\nFilled (0):\n", df_na.fillna(0))
print("\nInterpolated:\n", df_na.interpolate())
print("\nDropped Rows:\n", df_na.dropna())
```

## 6. Analysis: Grouping and Pivoting
Summarizing data for insights.

```python
# Re-creating a larger sample
df_analysis = pd.DataFrame({
    'Dept': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Level': ['Junior', 'Senior', 'Senior', 'Junior', 'Senior'],
    'Salary': [2000, 4000, 4500, 2500, 5000],
    'Experience': [1, 5, 6, 2, 8]
})

# Grouping
print("Mean Salary by Dept:\n", df_analysis.groupby('Dept')['Salary'].mean())
print("Count per Dept:\n", df_analysis.groupby('Dept').size())

# Complex Aggregation
agg_custom = df_analysis.groupby('Dept').agg({'Salary': ['min', 'max'], 'Experience': 'mean'})
print("\nCustom Agg:\n", agg_custom)

# Pivot Table
pivot = df_analysis.pivot_table(index='Dept', columns='Level', values='Salary', aggfunc='mean')
print("\nPivot Table:\n", pivot)
```
