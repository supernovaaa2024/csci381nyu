```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Find the dataset whether it is in the original exercise path or beside the notebook.
# for candidate in (Path('../data/gold_hist.txt'), Path('gold_hist.txt')):
#     if candidate.exists():
#         data_path = candidate
#         print(f'Using {candidate}')
#         break
# else:
#     raise FileNotFoundError('gold_hist.txt was not found in ../data/ or the notebook folder.')

raw = pd.DataFrame({
    'value': [line.strip() for line in open('gold_hist.txt').readlines() if line.strip()]
})
print(raw.head())
gold_hist = pd.DataFrame({
    'Year': raw.iloc[::2, 0].astype(int).to_numpy(),
    'Gold Price': (
        raw.iloc[1::2, 0]
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .astype(float)
        .to_numpy()
    )
}).sort_values('Year')



gold_hist_filtered = gold_hist[(gold_hist['Year'] >= 1972) & (gold_hist['Year'] <= 2015)].reset_index(drop=True)

plt.figure(figsize=(12, 6))
plt.plot(gold_hist_filtered['Year'], gold_hist_filtered['Gold Price'], marker='o', linestyle='-')
plt.title('Historical Gold Prices (1972-2015)')
plt.xlabel('Year')
plt.ylabel('Gold Price (USD per ounce)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
peak_price = gold_hist_filtered['Gold Price'].max()
peak_year = gold_hist_filtered.loc[gold_hist_filtered['Gold Price'] == peak_price]['Year'].values[0]
print(f"The peak gold price between 1972 and 2015 was ${peak_price:.2f} in the year {peak_year}.")

# Find the peak row
peak_idx = gold_hist_filtered['Gold Price'].idxmax()
peak_year = gold_hist_filtered.loc[peak_idx, 'Year']
peak_price = gold_hist_filtered.loc[peak_idx, 'Gold Price']
print(f"The peak gold price between 1972 and 2015 was ${peak_price:.2f} in the year {peak_year}.")
```
