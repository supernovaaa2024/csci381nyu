## **Data Structures**

### 1. Lists

A **list** in Python is an ordered, mutable collection used to store multiple items in a single variable. Lists can contain elements of different data types (such as integers, floats, strings, or even other lists), and their contents can be modified after creation.

Lists support indexing, slicing, iteration, and a wide range of built-in operations such as adding, removing, and searching for elements. Because of their flexibility and ease of use, lists are one of the most commonly used data structures in Python.


```python
lst = [1, 2, 3, "Hi", 5,6,9]            # create a list with initial elements
# lst.append(4)             # add 4 to the end of the list

# lst + [5]                 # concatenate list with [5]
# lst * 3                   # repeat the list twice
# 2 in lst                  # check if 2 exists in the list
# len(lst)                  # get the length of the list
# lst[1]                    # access element at index 1
# lst[0:5:2]                # slice the list from index 0 to 2 with step 1
# lst[::-1]
lst.append(2)             # add another 2 to the list
lst.count(2)              # count how many times 2 appears in the list
lst.index("Hi")              # get index of the first occurrence of 3
del lst[0]                # delete the element at index 0
lst
```

```python
lst = [1, 2, 3]
lst = lst + [5, 6]
print(lst)
# new_lst is [1, 2, 3, 5, 6]
# lst is still [1, 2, 3]
```

### 2. Tuples

A **tuple** is an ordered, **immutable** collection of elements in Python. Tuples are similar to lists, but once created, their values **cannot be changed**.


```python
t = (1, 2, 3, "Hi", True)        # create a tuple with initial elements
# t + (4,)            # concatenate tuple with (4,)
# t * 2               # repeat the tuple twice
# 2 in t              # check if 2 exists in the tuple
# len(t)              # get the length of the tuple
# t[1]                # access element at index 1
t[::2]              # slice the tuple from index 0 to 2 t[start:end:step]
# t
```

### 3. Sets
A **set** in Python is an **unordered collection of unique elements**.  
This means:
- Elements **cannot be duplicated**
- The order of elements is **not guaranteed**
- Sets are **mutable** (you can add or remove elements)

Sets are commonly used for **membership testing**, **removing duplicates**, and **mathematical set operations**.


```python
a = {1, 2, 3, "hasan", 3, "hasan", 1, "hasan"}     # define set a
b = {3, 4, 5}     # define set b

# a & b             # intersection of sets a and b
# a | b             # union of sets a and b
# a - b             # difference (elements in a but not in b)
a ^ b             # symmetric difference (elements in either set, but not both)
```

### 4. Dictionaries

A **dictionary** in Python is a built-in data type used to store data in **key–value pairs**.  
Each key maps to a value, allowing fast lookup, insertion, and deletion.

Dictionaries are:
- **Unordered** (before Python 3.7) / **Insertion-ordered** (Python 3.7+)
- **Mutable** (can be changed after creation)
- Indexed by **keys**, not by position


```python
# d = {"james": [100,200,3000], "maya": {"jan":200, "sep":300}, ("hasan", 123): 5000}    # create a dictionary with key-value pairs
# d["maya"]                 # access the value associated with key "a"
# list(d.keys())               # get all keys in the dictionary
# list(d.values())             # get all values in the dictionary
d={"a":100, "b":200}
# list(d.items())              # get all key-value pairs
# "a" in d               # check if key "a" exists in the dictionary
# del d["a"]             # delete the key "a" from the dictionary
# d["maya"] ["sep"]
# d
```

### 5. Numpy

- Provides the ndarray, a fast, efficient homogeneous array for numerical data.
- Ideal for mathematical operations, vectorized computations, and linear algebra.

Features:

- Fixed-size arrays with fast indexing and slicing
- Efficient memory usage for large datasets
- Supports broadcasting and element-wise operations

```python
import numpy as np

# Create arrays
arr1 = np.array([1, 2, 3, 4])             # 1D array
arr2 = np.array([[1, 2], [3, 4]])         # 2D array

# # Basic arithmetic
# print(arr1 + 10)                           # Add 10 to each element
# print(arr1 * 2)                            # Multiply each element by 2
# print(arr1 - np.array([1,1,1,1]))          # Element-wise subtraction


# # Array operations
# print(arr1.sum())                           # Sum of all elements
# print(arr1.mean())                          # Mean of elements
# print(arr1.max(), arr1.min())               # Max and Min
# print(arr1.argmax())                        # Index of max element

# # Indexing and slicing
# print(arr1[0])                              # First element
# print(arr1[1:3])                            # Slice: elements at index 1 and 2
# print(arr2[:, 1])                           # Second column of 2D array

arr1 = np.array([1, 2, 3, 4])             # 1D array

# # Reshape and size
# print(arr2.shape)                           # Shape of array (rows, columns)
# print(arr2.reshape(2,2))                    # Reshape to 4x1 array
# print(arr2.size)                            # Total number of elements
```

```python

# # Broadcasting
print(arr2 + np.array([10, 20]))            # Add row-wise (broadcasting)
print(arr2 * 2)                             # Multiply all elements by 2
```

```python
# # Logical operations
arr1 = np.array([1, 2, 3, 4]) 
# print(arr1 > 2)                             # Element-wise comparison -> boolean array
print(arr1[arr1 > 2])                        # Filter elements > 2
```

```python
# # Linear algebra (dot product)
arr3 = np.array([[1,2],[3,4]])
arr4 = np.array([[5,6],[7,8]])
print(np.dot(arr3, arr4))                   # Matrix multiplication
```

```python
# # Random numbers
rand_arr = np.random.rand(3,3)             # 3x3 array with random floats [0,1)
print(rand_arr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# relplot creates the grid and draws the plots in one step
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="day", kind="scatter")

plt.show()
```

```python
# 1. Initialize the grid object
g = sns.FacetGrid(tips, col="time", hue="day")

# 2. Manually map the plotting function
g.map(sns.scatterplot, "total_bill", "tip")

# 3. Manually add the legend
g.add_legend()

plt.show()
```
