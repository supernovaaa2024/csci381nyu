## Exercise 01

Write code to read data from: https://data.giss.nasa.gov/gistemp/tabledata_v3/GLB.Ts+dSST.txt . (You can use data from the file `GLB.Ts+dSST.csv` instead)

Keep all monthly temperatures (from Jan to Dec) between 1980 and 2015 ONLY (ignore any other data), then:

- Find minimum temperature in Jan between 1980 - 2015
- Find average temperature in each year between 1980 - 2015
- Find temperature in Jan 2000
- Find temperature in Jun 2009

```python
# Find minimum temperature in Jan between 1980 - 2015
import pandas as pd

monthly_columns = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

nasa = pd.read_csv("GLB.Ts+dSST.csv", skiprows=1)
nasa[monthly_columns] = nasa[monthly_columns].apply(pd.to_numeric, errors="coerce")

temps_1980to2015 = nasa.loc[nasa["Year"].between(1980, 2015), ["Year", *monthly_columns]].reset_index(drop=True)

min_temp = temps_1980to2015["Jan"].min()
print(f"The minimum temperature in January between 1980 and 2015 is {float(min_temp)}")
```

```python
# Find average temperature in each year between 1980 - 2015
average_temperature_yearly = temps_1980to2015.copy()
average_temperature_yearly["Average"] = average_temperature_yearly[monthly_columns].mean(axis=1)

print(f"Average temperature each year between 1980 and 2015 is:\n{average_temperature_yearly[['Year', 'Average']]}")
```

```python
# Find temperature in Jan 2000
Jan2000Temp = float(temps_1980to2015.loc[temps_1980to2015["Year"] == 2000, "Jan"].iloc[0])
print(f"Temperature in January 2000 is {Jan2000Temp}")
```

```python
# Find temperature in Jun 2009
Jun2009Temp = float(temps_1980to2015.loc[temps_1980to2015["Year"] == 2009, "Jun"].iloc[0])
print(f"Temperature in June 2009 is {Jun2009Temp}")
```

## Extra Credit  (Optional)

`answers.txt` contains students answers of exam questions (each line contains student_id, then answers made by that student)

- The first number is the student’s ID number. The following 25 letters are the student responses to the exam. If there is no letter following a comma, this means the student skipped answering the question.

- Write code to read the data from the file, compare students answers to answers key (1 point for each correct answer), then find: 

    - The grade for each student.
    - Id of student of the highest score.
    - Id of student of the lowest score.

`answerkey = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"`

```python
answerkey = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
```

```python
# The grade for each student
correct_answers = answerkey.split(",")
student_answers = pd.read_csv("answers.txt", header=None, keep_default_na=False)

student_answers.columns = ["student_id", *[f"Q{i}" for i in range(1, 26)]]
responses = student_answers.iloc[:, 1:]
grades = student_answers[["student_id"]].copy()
grades["score"] = responses.eq(correct_answers).sum(axis=1)


print(f"Grades for each student are:\n{grades}")
```

```python
# Id of student of the highest score
id_highest = grades.loc[grades["score"].idxmax(), "student_id"]
print(f"ID of student with the highest score is {id_highest}")
```

```python
# Id of student of the lowest score
id_lowest = grades.loc[grades["score"].idxmin(), "student_id"]
print(f"ID of student with the lowest score is {id_lowest}")
```
