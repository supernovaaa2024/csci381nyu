## **Exercise 1**

### **Scenario**

- You consume data from multiple public APIs and want to ensure schema consistency before building a pipeline.

### **Data Source (Real API)**

GitHub public repositories API (no key needed):

- https://api.github.com/repos/pandas-dev/pandas
- https://api.github.com/repos/numpy/numpy
- https://api.github.com/repos/scikit-learn/scikit-learn

### **Tasks**

- Fetch JSON from each API.
- Verify whether there is any difference between the schemas (key sets)

```python
# Your python code here

import requests

urls = [
    "https://api.github.com/repos/pandas-dev/pandas",
    "https://api.github.com/repos/numpy/numpy",
    "https://api.github.com/repos/scikit-learn/scikit-learn"
]

# responses = list(map(lambda u: requests.get(u).json(), urls))

k0=set(requests.get(urls[0]).json().keys())
k1=set(requests.get(urls[1]).json().keys())
k2=set(requests.get(urls[2]).json().keys())
(k0 | k1 | k2) != (k0 & k1 & k2)
```
