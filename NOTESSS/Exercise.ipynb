```python
import pandas as pd
df=pd.read_csv('../data/countries.csv', sep=';')
df
```

```python
# keep the following columns only
#     ['Country','Continent','Population','Area','Coastline','Currency','Birthrate','Deathrate','Life expectancy']]
df=df[['Country','Continent','Population','Area','Coastline','Currency','Birthrate','Deathrate','Life expectancy']]
df
```

```python
# change the column names to lowercase
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
# drop columns with missing values
df.isna().sum().sum()
df=df.dropna()
df.head()
```

```python
# limit to countries with populations over 20 million
df=df.loc[df.population>20e6,:]
```

```python
# reset index to start from zero and count over remaining countries
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
# add a boolean valued column that is true for countries with coastlines
df['coastal'] = df['coastline']>0
df.head()
```

```python
# Find min and max of the 'life expectancy' for coastal & landlocked countries.
# df.groupby('coastal').agg(['min', 'max'])['life expectancy']
df[['life expectancy','coastal']].groupby('coastal').agg(['min', 'max'])
```

```python
# How many coastal and landlocked countries are in each continent? 
df.groupby(['coastal','continent']).size()
```

```python
df.groupby(['continent','coastal']).size()
```

```python
# Determine the mean deathrate for coastal and landlocked countries in each continent.
df[['continent','coastal','deathrate']].groupby(['coastal','continent']).mean()
```

```python
df.head()
```

```python
# Create a new column measuring population density. 
df['density']= df['population']/df['area']
df.head()
```

```python
 # Determine the mean population density of all the countries.
df.density.mean()
```

```python
# Create a new column 'dense' that is True for countries with higher than average density and False for lower than average density.
df['dense'] = df['density']>142.1239784002992
df.head()
```

```python
# For each category of 'dense' compute the mean birthrate and the difference between the maximal birthrate and the minimal birthrate.
def my_diff(x):
    return x.max() - x.min()
df[['birthrate','dense']].groupby('dense').agg(['mean',my_diff])
```
