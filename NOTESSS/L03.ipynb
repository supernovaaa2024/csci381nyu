- Numeric:

  - 2 = write
  - 1 = execute

- 7 = rwx
- 5 = r-x

## 2. Executing Scripts

```bash
echo -e "#!/bin/bash\necho Hello World" > run.sh   # Create a simple Bash script that prints "Hello World"
```

```bash
chmod +x run.sh       # Make the script executable
./run.sh              # Run the s
```

### 3. Accessing Script Arguments

```bash
$0                     # Name of the script
$1, $2, $3 ...         # First, second, third arguments
$#                     # Number of arguments passed
$@                     # All arguments as separate words
```

```bash
# Example:
echo "Script name: $0"
echo "First arg: $1, Second arg: $2"
echo "Total args: $#"
```

### 4. Arithmetic Expressions

```bash
x=$((a + b))       # Arithmetic using $(( ))
y=`expr $a + $b`   # Arithmetic using expr
(( x += 5 ))       # Increment x by 5
if (( x < 20 )); then echo "x is less than 20"; fi
```

### 5. String Expressions

```bash
if [[ $x == "james" ]]; then echo "Hi James"; fi
if [[ $x != "hasan" ]]; then echo "Not Hasan"; fi
```

### 6. Subshell / Command Substitution

```bash
x=$(cat data.txt)       # Capture output of a command
for i in $(cat file.txt); do echo $i; done
```

### 7. Conditionals

```bash
if [ $x -eq 9 ]; then
    echo "x is 9"
elif [ $x -lt 5 ]; then
    echo "x is less than 5"
else
    echo "x is something else"
fi
```

### 8. Loops

```bash
# for loop
for var in $@; do
    echo $var
done
```

```bash
# While Loop
while read line; do
    echo "Line: $line"
done < sample.txt
```

### 9. Comparison Operators

```bash
# Integer
-eq  # equal
-ne  # not equal
-gt  # greater than
-ge  # greater or equal
-lt  # less than
-le  # less or equal

# Boolean
-o   # OR
-a   # AND

# String
=    # equal
==   # equal
!=   # not equal
-z   # string length is zero
-n   # string length is not zero

# File Tests
-e FILE   # File exists
-d FILE   # Directory exists
-f FILE   # Regular file exists
-r FILE   # File is readable
-s FILE   # File has non-zero size
-w FILE   # File is writable
-x FILE   # File is executable
-u FILE   # Set-user-ID bit set
-g FILE   # Set-group-ID bit set
```
