# Mastering Data Visualization with Seaborn
This notebook provides a comprehensive guide to the Seaborn library, covering everything from distribution plots to categorical analysis and advanced faceting.

## 1. Setup and Sample Data
First, we import the necessary libraries and create a synthetic dataset for our examples.

```python
# !pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Setting the theme
sns.set_theme(style="whitegrid")

# Generating a sample dataset
np.random.seed(42)
data = {
    'age': np.random.normal(35, 10, 200).astype(int),
    'salary': np.random.normal(50000, 15000, 200),
    'height': np.random.normal(170, 10, 200),
    'weight': np.random.normal(70, 15, 200),
    'gender': np.random.choice(['Male', 'Female'], 200),
    'department': np.random.choice(['Sales', 'IT', 'HR', 'Marketing'], 200),
    'experience': np.random.randint(1, 25, 200),
    'performance_score': np.random.uniform(1, 10, 200)
}
df = pd.DataFrame(data)
df.head()
```

## 2. Distribution Plots
Used to visualize the spread and density of numeric data.

```python
# Histogram
sns.histplot(data=df, x="age", bins=20, kde=False);
```

```python
# Kernel Density Estimate (KDE)
sns.kdeplot(data=df, x="age", fill=True);
sns.kdeplot(data=df, x='age', fill=True, hue='gender');
```

```python
# Rug Plot
sns.rugplot(data=df, x="age");
```

```python
# ECDF Plot (Empirical Cumulative Distribution)
sns.ecdfplot(data=df, x="age");
```

## 3. Relational Plots
Used to identify relationships between multiple variables.

```python
# Scatter Plot
sns.scatterplot(data=df, x='age', y='salary');
sns.scatterplot(data=df, x='age', y='salary', hue='gender');
```

```python
# Regression Plot
sns.regplot(data=df, x="age", y="salary");
sns.lmplot(data=df, x='age', y='salary', hue='gender');
```

```python
# Line Plot
sns.lineplot(data=df, x='age', y='salary');
```

```python
# Pair Plot (Pairwise relationships)
sns.pairplot(df[['age', 'salary', 'experience', 'gender']], hue="gender");
```

```python
# Joint Plot (Scatter + Distribution)
sns.jointplot(data=df, x="height", y="weight");
sns.jointplot(data=df, x="height", y="weight", kind="hex");
```

## 4. Categorical Plots
Used to compare a numeric variable across different categories.

```python
# Count Plot
sns.countplot(data=df, x="department");
sns.countplot(data=df, x="department", hue="gender");
```

```python
# Box Plot
sns.boxplot(data=df, x="department", y="salary");
```

```python
# Violin Plot
sns.violinplot(data=df, x="gender", y="performance_score");
```

```python
# Strip plot
sns.stripplot(data=df, x='gender', y='salary');
```

```python
# Swarm Plot
sns.swarmplot(data=df, x="department", y="experience");
```

## 6. Faceting (Small Multiples)
Breaking down plots into a grid based on categorical variables.

```python
# FacetGrid Example
sns.FacetGrid(df, col="gender", row="department").map(sns.histplot, "age")
sns.FacetGrid(df, col="gender", row="department", hue='department').map(sns.histplot, "age")
# Seaborn does not have a 'scatter' function; use 'scatterplot' instead.
sns.FacetGrid(df, col="gender", row="department").map(sns.scatterplot, "age", "salary")
```

```python
# Categorical Wrapper (catplot)
sns.catplot(data=df, x="department", y="salary", kind="boxen");
sns.catplot(data=df, x="department", y="salary", col="gender", kind="boxen")
```
