## 9. Searching & Finding

```bash
find . -name "*.txt"   # Find all .txt files starting from the current directory
find . -type f         # Find all regular files starting from the current directory
find . -type d         # Find all directories starting from the current directory
```

```bash
grep "ERROR" sample.txt          # find all lines containing "ERROR"
grep -i "error" sample.txt        # case-insensitive search
grep -c "ERROR" sample.txt        # count matching lines
grep -n "ERROR" sample.txt        # show line numbers of matches
grep -v "INFO" sample.txt         # show lines NOT containing "INFO"
grep -r "TODO" ./data/            # recursive search in directory
grep -w "ERROR" sample.txt        # match whole word only
```

# 11. Redirection & Pipes

```bash
ls > files.txt            # Redirect ls output to files.txt (overwrite if it exists)
ls >> files.txt           # Append ls output to the end of files.txt
wc < files.txt            # Count lines, words, and bytes from files.txt via input redirection
cat files.txt | wc -l     # Count the number of lines in files.txt using a pipe
```

## 12. wc (Word Count)

```bash
wc sample.txt        # Show number of lines, words, and bytes in sample.txt
wc -l sample.txt     # Show number of lines in sample.txt
wc -w sample.txt     # Show number of words in sample.txt
wc -c sample.txt     # Show number of bytes in sample.txt
```

## 13. sort

```bash
echo -e "3\n1\n2" > nums.txt   # Create nums.txt with three lines: 3, 1, 2
```

```bash
sort nums.txt                  # Sort the lines alphabetically (lexicographically)
sort -r nums.txt               # Sort the lines in reverse order
sort -n nums.txt               # Sort the lines numerically
sort -b nums.txt               # Sort lines ignoring leading spaces
```

## 15. cut

```bash
echo "a:b:c:d" > cut.txt         # Create cut.txt with the text "a:b:c:d"
```

```bash
cut -c1-3 cut.txt                # Extract characters 1 to 3 from each line
cut -d: -f1 cut.txt              # Extract the first field using ':' as the delimiter
cut -d: -f1,3 cut.txt            # Extract the first and third fields separated by ':'
cut -d: -f-3 cut.txt             # Extract fields 1 through 3 (inclusive) separated by ':'
cut -d: -f1-3 cut.txt            # Another way to extract fields 1 through 3 separated by ':'
cut -b1-4 cut.txt                # Extract bytes 1 to 4 from each line
```

## 18. tr

```bash
echo "hello" | tr 'e' 'a'            # Replace 'e' with 'a', output: hallo
echo "java" | tr '[ja]' '[eb]'       # Replace 'j'→'e' and 'a'→'b', output: ebbb
echo "hello" | tr 'a-z' 'A-Z'        # Convert all lowercase letters to uppercase, output: HELLO
echo "xxxxy" | tr -s 'x'             # Squeeze repeated 'x' characters into a single 'x', output: xy
echo "abcx" | tr -d 'x'              # Delete 'x' characters, output: abc
echo "a 1 b 2" | tr -d '[:space:]'   # Delete all whitespace characters, output: a1b2
echo "a1b2" | tr -d '[:digit:]'      # Delete all digits, output: ab
echo "a1b2" | tr -cd '[:digit:]'     # Keep only digits, deleting everything else, output: 12
```

```bash
echo "hello" | tr "e" "a"
echo "hello" | tr "[he]" "[el]"
```

## 19. sed

```bash
echo -e "apple\nbanana\napple\nbacon\ncat\ndog\napple pie\na b   c\n123\nb\nx y z" > sample.txt
```

```bash
sed 's/apple/orange/' sample.txt       # Replace first occurrence of 'apple' with 'orange' per line
sed 's/apple/orange/g' sample.txt      # Replace all occurrences of 'apple' with 'orange' per line
sed '2s/apple/orange/' sample.txt      # Replace 'apple' with 'orange' only on line 2
sed '/apple/s/apple/orange/' sample.txt # Replace 'apple' with 'orange' only on lines containing 'apple'
sed 's/^/hi/' sample.txt                # Add 'hi' to the beginning of each line
sed 's/$/hi/' sample.txt                # Add 'hi' to the end of each line
```

```bash
sed '3d' sample.txt       # Delete line 3
sed '1,2d' sample.txt     # Delete lines 1 to 2
sed '$d' sample.txt       # Delete last line
sed '$d' sample.txt       # Delete last line
```

## 20. awk

```bash
echo "10 20" | awk '{print $0}'        # Print the entire line
echo "10 20" | awk '{print $1}'        # Print the first field
echo "10 20" | awk '{print $2}'        # Print the second field
echo "10 20" | awk '{print $1"\t"$2}'  # Print first and second fields separated by a tab
echo "10 20" | awk '{print $NF}'       # Print the last field of the line
```

```bash
awk "{print $0}" sample.txt
```

```bash
echo "10,20" | cut -d ,  -f1-2
# Can only extract, cannot calculate
```

```bash
echo "a:b:c" | awk -F ":" '{print $1}' # Print the first field using ':' as delimiter
echo "a:b:c" | awk -F ":" '{print $2}' # Print the second field using ':' as delimiter
```

```bash
echo "10 20" | awk '{print $1 + $2}'   # Print the sum of the first and second fields
echo "10 20" | awk '{print $1 * $2}'   # Print the product of the first and second fields
```
